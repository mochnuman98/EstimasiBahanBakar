# GasType Berdasarkan Jenis Bahan Bakar

```python
gastype = df.groupby('distance').count()[['gas_type']].sort_values(by='distance',ascending=True).reset_index()
gastype = gastype.rename(columns={'gas_type':'RonBahanBakar'})
```

# Speed

# Seleksi Fitur

```python
features = ['distance','speed','temp_inside','temp_outside','gas_type']
x = df[features]
y = df['consume']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

```python
input_data = np.array([[28,26,215,12,98]])

prediction = lr.predict(input_data)
print('Estimasi Konsumsi Bahan Bakar :', prediction)
```

# Save Model

```python
import pickle

filename = 'Estimasi_Konsumsi_Bahan_Bakar.sav'
pickle.dump(lr,open(filename,'wb'))
```
